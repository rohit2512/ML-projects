```python
import glob
#to separate dataset file paths and their corresponding classes
path = r"C:\Users\rnegi\Downloads\newsGroup\20_newsgroups"
x_files = glob.glob(path+"\*\**")
y_files = []
for i in range(len(x_files)):
    #to get the class name from the full file path
    y_files.append(x_files[i].split("\\")[6])
```

```python
#to separate training files and testing files
from sklearn import model_selection
x_tr, x_ts, y_tr, y_ts = model_selection.train_test_split(x_files,y_files,test_size = 0.3, random_state = 42)
```

```python
#this function go through each and every file and generates a list of words from that
def build_vocab(x_tr):
    l = []
    for i in range(len(x_tr)):
        with open(x_tr[i],'r') as f:
            file_content = f.readline()
            word_list = file_content.split()
            l.extend(word_list)
            while(len(file_content) > 0):
                file_content = f.readline()
                word_list = file_content.split()
                l.extend(word_list)   
    return l
```

```python
#this function removes the non-alphabetic words from the list generated by previous function
def remove_nonalphabetic_words(l):
    i = 0
    end = len(l)
    while(i < end):
        if l[i].isalpha():
            l[i] = l[i].lower()
            i += 1
        else:
            l.pop(i)
            end -= 1
    return l
```

```python
#code to convert the training data files in classifier feedable form that is into a ndnumpy.array
#size is LEN(DATA) cross LEN(VOCABULARY)
def convert_to_classifier_feedable_form(data,vocabulary):
    import numpy as np
    X = np.zeros((len(data),len(vocabulary)),int)
    print(X)
    i = 0
    for path in data:
        l = []
        with open(path, 'r') as f:
            file_content = f.readline()
            word_list = file_content.split()
            l.extend(word_list)
            while(len(file_content) > 0):
                file_content = f.readline()
                word_list = file_content.split()
                l.extend(word_list)  
        l1 = remove_nonalphabetic_words(l)
        l2 = remove_stopwords(l1)
        for word in l2:
            if word in vocabulary:
                X[i][vocabulary.index(word)] += 1
        i += 1
    return X
```

```python
#training and testing files are converted into ndnumpy.array containing the frequency of each word in a document if it belongs to vocabulary
X_TRAIN = convert_to_classifier_feedable_form(x_tr,vocabulary)
X_TEST = convert_to_classifier_feedable_form(x_ts,vocabulary)
```

# now the code for multinomial naive bayes:-

```python
import numpy as np
```

```python
#code to prepare the dictionary
#top level elements are the classes
#following level is the each and every feature
#then is the sum of values in each feature in rows belonging to a particular class
def training(X_train, Y_train):
    #empty dictionary
    dictionary = {}
    #to get list of distinct classes
    classes = set(Y_train)
    #total_count is for storing the number of data points in training datasets
    dictionary["total_count"] = len(Y_train)
    for current_class in classes:
        current_class_rows = (Y_train == current_class)
        x_current = X_train[current_class_rows]
        dictionary[current_class] = {}
        num_features = X_train.shape[1]
        #total_words_count is for storing the sum of frequencies of all words belonging to a particular class
        dictionary[current_class]["total_words_count"] = 0
        #length stores the number of data points belonging to that particular class
        dictionary[current_class]["length"] = len(x_current)
        for i in range(1, num_features+1):
            dictionary[current_class][i] = x_current[:,i-1].sum()
            dictionary[current_class]["total_words_count"] += dictionary[current_class][i]
    return dictionary
```

```python
def probability(x, dictionary, current_class):
    output = np.log(dictionary[current_class]["length"]) - np.log(dictionary["total_count"])
    num_features = len(dictionary[current_class].keys()) - 2
    for j in range(1, num_features+1):
        numerator = dictionary[current_class][j] + 1
        denominator = dictionary[current_class]["total_words_count"] + len(dictionary[current_class].keys()) - 2
        xj = x[j-1]
        current_xj_prob = 0
        for i in range(xj):
            current_xj_prob = current_xj_prob + np.log(numerator) - np.log(denominator)
        output = output + current_xj_prob
    return output
```

```python
def predict_single_point(x, dictionary):
    classes = dictionary.keys()
    p_best_class = -1000
    best_class = -1
    first_run = True
    for current_class in classes:
        if (current_class == "total_count"):
            continue
        p_current_class = probability(x, dictionary, current_class)
        if(first_run or p_current_class > p_best_class):
            p_best_class = p_current_class
            best_class = current_class
        first_run = False
    return best_class
```

```python
def testing(X_test, dictionary):
    y_pred = []
    for i in range(X_test.shape[0]):
        x = X_test[i][:]
        x_class = predict_single_point(x, dictionary)
        y_pred.append(x_class)
    return y_pred
```

# predictions from self written multinomial naive bayes

```python
dict = training(X_TRAIN, Y_TRAIN)
y_pred = testing(X_TEST, dict)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_TEST,y_pred))
from sklearn.metrics import classification_report
print(classification_report(Y_TEST,y_pred))
```
