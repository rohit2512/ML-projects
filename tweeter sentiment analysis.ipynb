```python
import pandas as pd
tweets = pd.read_csv(r"C:\Users\rnegi\Downloads\0000000000002747_training_twitter_x_y_train.csv")
test_data = pd.read_csv(r"C:\Users\rnegi\Downloads\0000000000002747_test_twitter_x_test.csv")
df = tweets
ds = test_data
```

```python
text = df['text']
text = list(text)
sentiment = df['airline_sentiment']
sentiment = list(sentiment)

test_text = ds['text']
test_text = list(test_text)
```

```python
from nltk import pos_tag
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
vocab  = []
for sentence in text:
    temp = []
    for word in sentence.split():
        if word.lower() not in stops:
            pos = pos_tag([word])
            clean_word = lemmatizer.lemmatize(word, pos = get_simple_pos(pos[0][1]))
            temp.append(clean_word)
    vocab.append(" ".join(temp))
```

```python
vocab1  = []
for sentence in test_text:
    temp = []
    for word in sentence.split():
        if word.lower() not in stops:
            pos = pos_tag([word])
            clean_word = lemmatizer.lemmatize(word, pos = get_simple_pos(pos[0][1]))
            temp.append(clean_word)
    vocab1.append(" ".join(temp))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vec = CountVectorizer(max_features = 2000, min_df = 0.0001, max_df = 0.999)
x_tr = count_vec.fit_transform(vocab)
x_ts = count_vec.transform(vocab1)
```

```python
count_vec.get_feature_names()
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(x_tr, sentiment)
y_pred = clf.predict(x_ts)

import numpy as np
np.savetxt(r"C:\Users\rnegi\Downloads\sentiment_output.csv", y_pred, fmt = '%s')
```
